# Titanic Mini Visualization Dashboard

## Overview
This notebook analyzes the Titanic dataset using Python.  
The goal is to clean the data, create simple new features, and build visualizations to understand survival patterns.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
```

## Load Dataset
Here, we load the Titanic CSV file and preview the first few rows.

```python


url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df.head()
```

```python
df.info()
```

## Data Cleaning
Missing values are handled by filling Age with median value and Embarked with mode value.  
The Cabin column is removed because it contains many missing values.

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df = df.drop(columns=["Cabin"], errors="ignore")

df.isnull().sum()
```

## Feature Engineering
Two new features are created:
- FamilySize: total family members onboard
- AgeGroup: passengers divided into age categories

```python
df["FamilySize"] = df["SibSp"] + df["Parch"]

df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0, 12, 18, 30, 50, 80],
    labels=["Child", "Teen", "YoungAdult", "Adult", "Senior"]
)

df.head()
```

## Visualization 1: Age Distribution
This chart shows the age distribution of Titanic passengers.

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=20, kde=True)
plt.title("Age Distribution of Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

### Insight
Most passengers were young adults, especially between 20 and 40 years old.

## Visualization 2: Survival Rate by Gender
This chart compares survival rate between male and female passengers.

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Sex", y="Survived", data=df)
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()
```

### Insight
Female passengers had a much higher survival rate than male passengers.

## Visualization 3: Fare Distribution by Passenger Class
This boxplot shows how ticket fare changes across passenger classes.

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="Pclass", y="Fare", data=df)
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

### Insight
First-class passengers paid higher fares compared to second and third class passengers.

## Visualization 4: Age vs Fare by Survival
This scatterplot shows the relationship between Age and Fare, colored by survival status.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df, alpha=0.7)
plt.title("Age vs Fare Colored by Survival")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.show()
```

### Insight
Passengers with higher fares show better survival chances, while age alone does not clearly separate survival.

## Visualization 5: Correlation Heatmap
This heatmap shows correlation between numeric columns.

```python
plt.figure(figsize=(8,6))
numeric_df = df.select_dtypes(include=["int64", "float64"])

sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### Insight
Survived is positively related to Fare and negatively related to Pclass.  
This means passengers in higher classes had better survival chances.

## Visualization 6: Survival Rate by Age Group
This chart shows survival rate across different age groups.

```python
plt.figure(figsize=(8,5))
sns.barplot(x="AgeGroup", y="Survived", data=df)
plt.title("Survival Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.show()
```

### Insight
Children show a relatively better survival rate compared to some older age groups.

## Conclusion
In this dashboard, Titanic data was cleaned and analyzed using pandas, matplotlib, and seaborn.  
The analysis shows that gender, passenger class, fare, and age group played important roles in survival chances.
